```python
def convert_duration(duration):
    # Separate decimal part for minutes
    hours = int(duration)
    minutes = round((duration - hours) * 60)
    
    return f"{hours} hours and {minutes} minutes"

# Example usage
duration_example = 2.17
print(convert_duration(duration_example))
data["duration"] = data["duration"].apply(convert_duration)
data["duration"]    


```

### 🌐 Observation: Destination-wise Ticket Price Trends per Source City

- Ticket prices vary **significantly based on destination**, even when the source city remains the same.
- For most source cities (like **Delhi**, **Mumbai**, **Kolkata**), flights to **Chennai** and **Kolkata** tend to be more expensive compared to other destinations.
- From **Bangalore** and **Hyderabad**, flights to **Chennai** show a noticeable price spike, suggesting high demand or limited availability on this route.
- Flights within the same city (e.g., Delhi to Delhi) show the lowest prices, indicating data placeholders or roundtrip/mirroring in the dataset.
- Overall, this chart highlights that **flight pricing is heavily route-dependent**, and strategic planning of source-destination pairs can impact travel costs significantly.

- This indicates that **early booking** leads to more economical ticket prices, while **last-minute bookings** are costlier.

```python
# Top 10 expensive route 
route_price = data.groupby(["source_city","destination_city"])["price"].mean().sort_values(ascending = False)
```

```python
plt.figure(figsize=(18,6))
route_price[:10].plot(kind = "bar",color = "#E2BF23",label = "Top 10 Expensive Route")
plt.title("Top 10 Expensive Route (By Avarege Price)")
plt.xlabel("Route")
plt.ylabel("Average Price")

plt.legend()
plt.show()
```

### 💸 Observation: Top 10 Most Expensive Flight Routes (by Average Price)

- The **Chennai–Bangalore** route has the **highest average ticket price**, exceeding ₹25,000.
- Routes involving **Chennai**, **Bangalore**, and **Kolkata** appear multiple times in the top 10, suggesting higher average fares for flights among these cities.
- All top 10 routes have average prices in the range of **₹22,500 to ₹25,000**, indicating relatively consistent pricing among the costliest routes.
- These routes may be premium due to higher demand, limited availability, or business-travel dominance.
